## Sintom√°ticos

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
import glob
fn = []
all_spectra = pd.DataFrame()

for f in glob.glob("./S/*.txt"):
    df = pd.read_csv(f, header=None, delimiter=' ' )
    all_spectra = pd.concat([all_spectra, df], axis=1)
    fn.append(f) 
Datamz = all_spectra[0] # realciones m/z
Data = all_spectra[1] # Intensidades
```

```python
Dataint=Data[0:23040]
Dataint
```

```python
xs = np.array(Datamz)
ys = np.array(Dataint)

fig, esp= plt.subplots(1,2, figsize=(20, 6))
plt.rcParams["font.family"] = 'Times New Roman'
esp[0].plot(xs, ys)

esp[0].set_title("Todas las muestras Sintom√°ticas", fontsize=20)
esp[0].set_xlabel("m/z", fontsize=20)
esp[0].set_ylabel("Relative intensity %", fontsize=20)

esp[0].set_xlim([570, 5800])
esp[1].set_xlim([570, 5800])

esp[1].plot(xs, ys.mean(axis=1), color='#FF9843', label='espectro promedio')

esp[1].set_title("Media de las muestras Sintom√°ticas", fontsize=20)
esp[1].set_xlabel("m/z", fontsize=20)
esp[1].set_ylabel("Relative intensity %", fontsize=20)

fig.savefig("sintomaticos.jpg", dpi=200)
plt.show()
```

## Asintom√°ticas

# Matriz
